```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from os import path
```

```python
file_name = path.join(path.curdir,'../data/external/countries.csv')
df = pd.read_csv(file_name)
```

```python
df.head()
```

```python
data_2007 = df[df.year==2007]
```

```python
plt.scatter(data_2007.gdpPerCapita, data_2007.lifeExpectancy)
plt.show()
```

```python
plt.scatter(np.log10(data_2007.gdpPerCapita), data_2007.lifeExpectancy)
plt.show()
```

```python
print('Life expectancy correlation with FDP (absolute) : {}'.format(data_2007.gdpPerCapita.corr(data_2007.lifeExpectancy)))
print('Life expectancy correlation with FDP (log10) : {}'.format(np.log10(data_2007.gdpPerCapita).corr(data_2007.lifeExpectancy)))
```

```python
years_sorted = sorted(set(df.year))
years_sorted
```

```python
for yr in years_sorted:
    data = df[df.year==yr]
    plt.scatter(np.log10(data.gdpPerCapita), data.lifeExpectancy,10)
    plt.title('Year {}'.format(yr))
    plt.xlabel('GDP')
    plt.ylim(25,85)
    plt.xlim(2,5)
    plt.ylabel('Life expectancy')
    plt.show()
```

```python
base_path=path.join(path.curdir,'../data/temp')
```

```python
for yr in years_sorted:
    data = df[df.year==yr]
    plt.scatter(np.log10(data.gdpPerCapita), data.lifeExpectancy,10)
    plt.title('Year {}'.format(yr))
    plt.xlabel('GDP')
    plt.xlim(0,5)
    plt.ylabel('Life expectancy')
    filename = path.join(base_path,str(yr))
    plt.savefig(filename, dpi=200)
    plt.clf()
```
