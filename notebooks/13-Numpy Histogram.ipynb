```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.histogram([1,2,1], bins=[0,1,2,3])
```

```python
mean,stddev = 2, 0.5
data = np.random.normal(mean,stddev,10000)
```

```python
n, bin_edges = np.histogram(data, bins=50)
```

```python
plt.plot(bin_edges[1:],n)
plt.show()
```
