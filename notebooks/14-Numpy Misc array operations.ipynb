```python
import numpy as np
from os import path
```

```python
file_name = path.join(path.curdir,'../data/external/salary.csv')
salaries = np.genfromtxt(file_name,delimiter=',')
```

```python
# Index of maximum salary
max_index = np.argmax(salaries)
salaries[max_index]
# Index of minimum salary
min_index = np.argmin(salaries)
salaries[min_index]
```

```python
# get the sorted values indices
sorted_indices = np.argsort(salaries)[::-1]
salaries[sorted_indices]
```

```python
# get the indices by applying the filter
gt100 = np.where(salaries>100000)
salaries[gt100]
```

```python
len(gt100[0])
```

```python
condition = salaries > np.mean(salaries)
salaries[condition]
```

```python
np.extract(condition,salaries)
```
