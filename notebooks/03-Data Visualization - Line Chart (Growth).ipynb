```python
import pandas as pd
from matplotlib import pyplot as plt
from os import path
```

```python
file_name = path.join(path.curdir,'../data/external/countries.csv')
df = pd.read_csv(file_name)
```

```python
china = df[df.country=='China']
us = df[df.country=='United States']
us.head()
```

```python
plt.plot(us.year, us.gdpPerCapita)
plt.title('GDP Per Capita Since 1950')
plt.plot(china.year, china.gdpPerCapita)
plt.legend(['United States','China'])
plt.xlabel('Year')
plt.ylabel('GDP Per Capita')
plt.show()
```

```python
china_growth = (china.gdpPerCapita/china.gdpPerCapita.iloc[0])*100
us_growth = (us.gdpPerCapita/us.gdpPerCapita.iloc[0])*100
```

```python
plt.plot(us.year, us_growth)
plt.title('GDP Per Capita Growth Since 1950')
plt.plot(china.year, china_growth)
plt.legend(['United States','China'])
plt.xlabel('Year')
plt.ylabel('GDP Per Capita')
plt.show()
```
