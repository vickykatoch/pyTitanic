```python
import pandas as pd
from pandas import isnull
import numpy as np
from os import path
```

```python
def mem_mib(df):
    print("{0:.2f} Mib".format(df.memory_usage().sum()/(1024*1024)))

def make_categorical(df, col_name):
    df.loc[:, col_name] = pd.Categorical(df[col_name])
    
def recent_nhl_only(df):
    return df[(df["lgID"] == "NHL") & (df["year"] >= 1980)]
```

```python
file_name = path.join(path.curdir,'../data/external/advpandas/TeamSplits.csv')
teamSplits = pd.read_csv(file_name)
recent_nhl_only(teamSplits)
teamSplits.head()
```

```python
cols_to_drop = teamSplits.columns[3:11]
teamSplits = teamSplits.drop(columns = cols_to_drop)
teamSplits = teamSplits.drop(columns = "lgID")
make_categorical(teamSplits,"tmID")
```

```python
p_file_name = path.join(path.curdir,'../data/external/advpandas/processed/TeamSplits.pickle')
teamSplits.to_pickle(p_file_name)
```
