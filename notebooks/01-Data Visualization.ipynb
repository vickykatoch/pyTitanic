```python
import pandas as pd
from matplotlib import pyplot as plt
from os import path
```

```python
df = pd.DataFrame({
    "year" : [2008,2012,2016],
    "age" : [24, 32, 36],
    "attendees" : [100,187, 201]
})
```

```python
plt.plot(df.year, df.attendees)
plt.plot(df.year, df.age)
plt.xlabel('Year')
plt.legend(['Attendees', 'Avg Age'])
plt.plot()
```
